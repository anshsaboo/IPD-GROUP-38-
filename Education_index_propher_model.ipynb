```python
import pandas as pd
import numpy as np
from prophet import Prophet
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
file_path = "/content/education_index_wards_mumbai_template.xlsx"
data = pd.read_excel(file_path,sheet_name="Sheet1")
```

```python
n_months = 36
date_range = pd.date_range(start="2020-01-01",periods=n_months,freq="M")
```

```python
simulated_data = []
```

```python
for _, row in data.iterrows():
  ward = row["Ward Name"]
  base_value = row["Education Index"]

  growth_factor = (
      row["Literacy Rate"]*0.05
      -row["Dropout Rate"]*0.03
      +row["Digital Access"]*0.04
      +np.random.normal(0,0.01)
  )

  values = [base_value]
  for t in range(1,n_months):
    new_val = values[-1] + growth_factor + np.random.normal(0,0.005)
    new_val = max(0,min(1,new_val))
    values.append(new_val)

  for ds,y in zip(date_range,values):
    simulated_data.append([ward,ds,y])
```

```python
ts_df=pd.DataFrame(simulated_data,columns=["ward","ds","y"])

```

```python
results = []
```

```python
wards = ts_df["ward"].unique()
```

```python
for ward in wards:
  ward_data = ts_df[ts_df["ward"]==ward][["ds","y"]]
  # Split into train/test: reserve last 6 months for testing
  train = ward_data.iloc[:-6]    # All but last 6 rows
  test = ward_data.iloc[-6:]     # Last 6 rows

  model = Prophet(
      daily_seasonality=False,
      weekly_seasonality=False,
      yearly_seasonality=True
  )

  model.fit(train)
  future = model.make_future_dataframe(periods=6,freq="M")
  forecast = model.predict(future)
  preds = forecast.iloc[-6:]["yhat"].values #yhat is prophet's prediction column

  mae = mean_absolute_error(test["y"].values,preds)
  rmse = np.sqrt(mean_squared_error(test["y"].values, preds))

  results.append([ward, mae, rmse])

  plt.figure(figsize=(10, 6))
  plt.plot(train["ds"], train["y"], label="Train Data", color="blue")
  plt.plot(test["ds"], test["y"], label="Actual Test Data", color="black", marker="o")
  plt.plot(test["ds"], preds, label="Forecast", color="red", linestyle="--", marker="x")

  plt.title(f"Ward {ward} Forecasting\nMAE={mae:.4f}, RMSE={rmse:.4f}")
  plt.xlabel("Date")
  plt.ylabel("Education Index")
  plt.legend()
  plt.grid(True)
  plt.show()


```

```python
results_df = pd.DataFrame(results,columns = ["Ward","MAE","RMSE"])
```

```python
results_df.loc[len(results_df)] = [
    "Average",
    results_df["MAE"].mean(),
    results_df["RMSE"].mean()
]
```

```python
print("\nProphet Forecasting Errors per Ward:")
print(results_df)
```
